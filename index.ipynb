# **Extração**

```python
url = 'https://api.coinbase.com/v2/prices/spot'
response = requests.get(url)
response = response.json()['data'] # Quantia, Qual e a cripto meoda, Moeda
print(response)
```

# **Trasformação**

```python
moeda = response['currency']
criptomoeda = response['base']
valor = response['amount']

dados_transformandos = {
    'Valor': valor,
    'Criptomoeda': criptomoeda,
    'Moeda': moeda
}

dados_transformandos
```

```python
d = 1
data = []
while d < 40:
    print(dados_transformandos)
    data.append(dados_transformandos)
    d+= 1
```

```python
df = pd.DataFrame(data)
df
```
